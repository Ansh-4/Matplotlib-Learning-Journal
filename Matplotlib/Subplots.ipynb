# Subplots
Subplots in Matplotlib let you display multiple plots in a single figure, arranged in rows and columns.
They are extremely useful in EDA when you want to compare multiple features at once.

## Figure
the entire canvas

## Ax
A single plot (subplot) //Axes are technically numpy arrays

```python
#print(plt.subplot(2,2)) #rows and columns
```

This code returns a tuple. The tuple has two entities a figure object which is our canvas and a 2-d array of axes objects. When calling this function we are going to unpack this tuple right away

```python
figure, axes = plt.subplots(2,2)
plt.show()
```

```python
figure, axes = plt.subplots(1,2)
plt.show()
```

```python
figure, axes = plt.subplots(2,1)
plt.show()
```

## Filling data

```python
figure, axes = plt.subplots(2,2)

x=np.array([1,2,3,4,5])

axes[0,0].plot(x, x*2, color="red")
axes[0,0].set_title("x*2")

axes[0,1].plot(x, x**2, color="blue")
axes[0,1].set_title("x**2")

axes[1,0].plot(x,x**3,color="yellow")
axes[1,0].set_title("x**3")

axes[1,1].plot(x, x**4, color="green")
axes[1,1].set_title("x**4")

plt.show()
```

```python
figure, axes = plt.subplots(2,2)

x=np.array([1,2,3,4,5])

axes[0,0].plot(x, x*2, color="red")
axes[0,0].set_title("x*2")

axes[0,1].plot(x, x**2, color="blue")
axes[0,1].set_title("x**2")

axes[1,0].plot(x,x**3,color="yellow")
axes[1,0].set_title("x**3")

axes[1,1].plot(x, x**4, color="green")
axes[1,1].set_title("x**4")

plt.tight_layout()
plt.show()
```

## We can also use different charts

```python
figure, axes = plt.subplots(2,2)

x=np.array([1,2,3,4,5])

axes[0,0].plot(x, x*2, color="red")
axes[0,0].set_title("x*2")

axes[0,1].bar(x, x**2, color="blue")
axes[0,1].set_title("x**2")

axes[1,0].bar(x,x**3,color="yellow")
axes[1,0].set_title("x**3")

axes[1,1].plot(x, x**4, color="green")
axes[1,1].set_title("x**4")

plt.tight_layout()
plt.show()
```
