```python
import matplotlib.pyplot as plt
import numpy as np
```

# Pie Chart in Matplotlib — 

A pie chart is a circular chart used to show proportions or percentages of a whole.
Each slice (sector) represents a category’s contribution to the total.

```python
Year = ["First-Year", "Second-Year", "Third-Year", "Fourth-Year"] #these will be our labels for the pie-chart
Students = np.array([440, 400, 340, 200])

plt.pie(Students, labels=Year)

plt.show()
```

# Customizing the Pie-Chart

```python
Year = ["First-Year", "Second-Year", "Third-Year", "Fourth-Year"] #these will be our labels for the pie-chart
Students = np.array([440, 400, 340, 200])
colors = ["red", "yellow", "green", "blue"]

plt.pie(Students, labels=Year,
       colors=colors)

plt.show()
```

```python
Year = ["First-Year", "Second-Year", "Third-Year", "Fourth-Year"] #these will be our labels for the pie-chart
Students = np.array([440, 400, 340, 200])
colors = ["red", "yellow", "green", "blue"]

plt.pie(Students, labels=Year,
       colors=colors,
       autopct="%1.1f%%")

plt.show()
```

```python
Year = ["First-Year", "Second-Year", "Third-Year", "Fourth-Year"] #these will be our labels for the pie-chart
Students = np.array([440, 400, 340, 200])
colors = ["pink", "#6ec9f3", "#952eff", "#dc79e9"]

plt.pie(Students, labels=Year,
       colors=colors,
       autopct="%1.1f%%",
       explode=[0.1,0,0,0],
       shadow=True)

plt.show()
```

```python
Year = ["First-Year", "Second-Year", "Third-Year", "Fourth-Year"] #these will be our labels for the pie-chart
Students = np.array([440, 400, 340, 200])
colors = ["pink", "#6ec9f3", "#952eff", "#dc79e9"]

plt.title("College", fontsize=20, family="Arial", color="#ed680e")

plt.pie(Students, labels=Year,
       colors=colors,
       autopct="%1.1f%%",
       explode=[0.1,0,0,0],
       shadow=True, startangle=180)

plt.show()
```
