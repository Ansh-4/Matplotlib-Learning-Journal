# Bar Chart in Matplotlib —

A bar chart is a graph that uses rectangular bars to show and compare values of different categories.
The length/height of each bar is proportional to the value it represents.

## When to use a bar chart

Use a bar chart when:

Data is categorical (names, types, groups)

You want to compare quantities

Exact values matter more than trends

## Example use cases:

Marks of students

Sales per product

Population of cities

Votes per party

## Basic syntax

plt.bar(x, height)


x → categories (names or positions)

height → values for each category


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
categories = np.array(["Grains", "Fruits", "Vegetable", "Protein", "Dairy", "Sweets"])
values = np.array([3,4,6,5,3,1])

plt.bar(categories, values)

plt.show()
```

## Adding titles and Labels to our Bar chart

```python
categories = np.array(["Grains", "Fruits", "Vegetable", "Protein", "Dairy", "Sweets"])
values = np.array([3,4,6,5,3,1])

plt.bar(categories, values)

plt.title("Daily Consumption")
plt.xlabel("Food")
plt.ylabel("Quantity")

plt.show()
```

## Changing the color of our Bar Chart

```python
categories = np.array(["Grains", "Fruits", "Vegetable", "Protein", "Dairy", "Sweets"])
values = np.array([3,4,6,5,3,1])

plt.bar(categories, values, color="pink")

plt.title("Daily Consumption")
plt.xlabel("Food")
plt.ylabel("Quantity")

plt.show()
```

# Horizontal Bar Chart
use barh funtion instead of bar 

```python
categories = np.array(["Grains", "Fruits", "Vegetable", "Protein", "Dairy", "Sweets"])
values = np.array([3,4,6,5,3,1])

plt.barh(categories, values, color="red")

plt.title("Daily Consumption")
plt.xlabel("Food")
plt.ylabel("Quantity")

plt.show()
```
