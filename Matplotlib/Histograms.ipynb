```python
import matplotlib.pyplot as plt
import numpy as np
```

# Histogram
A histogram shows the distribution of a single numerical variable by grouping values into bins (ranges) and counting how many values fall into each bin.
i.e A visual representation of the distribution of quantitative data. They group values into a bin(Intervals) and count how many fall in each range.

```python
scores = np.random.normal(loc=75, scale=10, size=100)
scores = np.clip(scores, 0, 100) #setting bounds -- minimum and the maximum

plt.hist(scores)

plt.show()

```

```python
scores = np.random.normal(loc=75, scale=10, size=100)
scores = np.clip(scores, 0, 100) #setting bounds -- minimum and the maximum

plt.hist(scores, bins=10)

plt.show()

```

## Customization


```python
scores = np.random.normal(loc=75, scale=10, size=100)
scores = np.clip(scores, 0, 100) #setting bounds -- minimum and the maximum

plt.hist(scores, bins=10,
        color="pink",
        edgecolor="black")

plt.show()
```

## Labels and Titles

```python
scores = np.random.normal(loc=75, scale=10, size=100)
scores = np.clip(scores, 0, 100) #setting bounds -- minimum and the maximum

plt.hist(scores, bins=10,
        color="pink",
        edgecolor="black")

plt.title("Exam Scores")
plt.xlabel("Scores")
plt.ylabel("No. of Students")

plt.show()
```
