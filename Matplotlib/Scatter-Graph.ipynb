# Scatter Graph
A scatter graph (scatter plot) shows the relationship between two numerical variables by plotting data points on a 2D plane.
In Matplotlib, scatter plots are created using plt.scatter().
(Helps to understand the correlation)
example - Study hours vs test scores


## What does a scatter graph show?

Each dot represents one observation:

X-axis → independent variable

Y-axis → dependent variable

Scatter plots help you:

Detect patterns & trends

See correlation (positive, negative, none)

Identify outliers

Compare groups (using color or size

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5, 7, 8]) #Hours studied
y = np.array([55, 60, 65, 72, 85, 90, 92, 98]) #Grades

plt.scatter(x,y)
plt.show()
```

## Adding Labels and Title to our scatter graph

```python
x = np.array([0, 1, 2, 3, 4, 5, 7, 8]) #Hours studied
y = np.array([55, 60, 65, 72, 85, 90, 92, 98]) #Grades

plt.scatter(x,y)

plt.xlabel("Hours Studied")
plt.ylabel("Grade")

plt.title("Test Scores")


plt.show()
```

## Customization

```python
x1 = np.array([0, 1, 2, 3, 4, 5, 7, 8]) #Hours studied
y1 = np.array([55, 60, 65, 72, 85, 90, 92, 98]) #Grades
x2 = np.array([0, 1, 2, 2, 3, 7, 8, 8]) 
y2 = np.array([50, 56, 65, 70, 75, 90, 92, 95])

plt.scatter(x1,y1, color="green",
           alpha=0.5,
           s=100,
           label="Class A")
    
plt.scatter(x2,y2, color="red",
           alpha=0.5,
           s=100,
           label="Class B")

plt.xlabel("Hours Studied")
plt.ylabel("Grade")

plt.title("Test Scores")

plt.legend()
plt.show()
```
