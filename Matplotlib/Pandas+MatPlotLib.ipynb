```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## pandas + matplotlib

```python
df = pd.read_csv(r"C:\Users\Ryuk\Jupyter\Matplotlib\Pokemon-150.csv")
print(df)
```

```python
type_count = df["Type1"].value_counts()
plt.barh(type_count.index, type_count.values)
```

```python
type_count = df["Type1"].value_counts(ascending=True)
plt.barh(type_count.index, type_count.values, color="Orange",edgecolor="black")
plt.title("Pokemon by Type-1")
plt.xlabel("Count")
plt.ylabel("type")
plt.tight_layout()
plt.show()
```
