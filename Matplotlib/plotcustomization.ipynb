# What is pyplot in Matplotlib?

## pyplot is a submodule of Matplotlib that provides a simple, MATLAB-like interface for creating plots.

```python
x = [2023, 2024, 2025, 2026, 2027] #year
y = [15, 14, 21, 19, 21] #class size of each year

#create a basic plot
plt.plot(x,y)
plt.show()
```

# plot() function (Matplotlib)

## The plot() function is used to draw line graphs in Matplotlib.
It is the most commonly used plotting function.

Basic syntax

plt.plot(x, y)


x → values on x-axis

y → values on y-axis

Draws a line connecting points

the default behaviour of plot is to start at zero and place a point on each increment of 1. (Here the increment is 0.5)

```python
plt.plot(y)
plt.show()
```

 ## We can instead use numpy arrays as they are faster than lists.
